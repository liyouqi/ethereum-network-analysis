# 4.4 MEV Community Detection

In this notebook, we detect **MEV-related communities** in the heterogeneous Ethereum graph.

We focus on:

- Communities centred around DEX routers
- Clusters of addresses that frequently interact with routers
- MEV candidate addresses grouped into structural clusters

The goal is to move from **individual MEV candidates** to
**MEV clusters** that may represent coordinated strategies or teams.
