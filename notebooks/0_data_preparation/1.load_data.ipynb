```python
token_tx = pd.read_csv(os.path.join(RAW_DATA_PATH, "token_transfers_filtered.csv"))
token_tx.head()
# token_tx.info()
# print(token_tx.isnull().sum())

'''token transfers
Columns:
    -transaction_hash: the hash of the transaction
    -from_address: the sender address
    -to_address: the receiver address
    -block_number: the block number containing the transaction
    -token_address: the address of the token contract
    -rao_value: the amount of tokens transferred (in the smallest unit of the token)
    -block_timestamp: the timestamp of the block containing the transaction
    -usd_value: the USD value of the token transfer at the time of the transaction
'''
```
