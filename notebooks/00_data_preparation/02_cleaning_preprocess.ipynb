## Description

In this phase, we clean the raw transaction and token transfer data by normalizing addresses, converting timestamps, and removing invalid or zero-value transactions. The cleaned data is then saved for further analysis.

We just cleaned two datasets:
- `transactions_filtered.csv` -> `transactions_clean.csv`
- `token_transfers_filtered.csv` -> `token_transfers_clean.csv`
  
Other two datasets is not nessary to clean, because `contracts_filtered.csv` is just address labels and used for reference searching.  `blocks_filtered.csv` is mapping block number to timestamp, used for time series analysis later.
